```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    use_safetensors=True
)
pipe = pipe.to("cuda")
```

```python
def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image
```

```python
interface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter your image prompt", placeholder="e.g. Shahrukh Khan as a superhero"),
    outputs="image",
    title="ðŸ§  AI Text-to-Image Generator",
    description="Enter a prompt and get an image generated by Stable Diffusion"
)
```
